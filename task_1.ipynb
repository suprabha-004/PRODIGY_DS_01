```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random age data
np.random.seed(500) # static 
ages = np.random.randint(1, 100, size=1000)  # Generating 1000 random ages between 1 and 100

# Create a histogram with improved visual representation
plt.hist(ages, bins=20, color='green', edgecolor='black', alpha=0.9)  # alpha for transparency, edgecolor for separation and bins for width
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages versus Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Add grid lines for enhanced visibility
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
